<h1>API References</h1>

<h2>
class Interface()
</h2>

<p>
This class identifies an interface connecting to an RTL/HLS block from another block.
</p>

<style>
    .funcbox{padding-left: 30px;}
</style>

<div class = "funcbox">
<h3>Interface.__init__(name, **kwargs)</h3>
<p>
This function initializes a port group object.
</p>
<table>
<tr><td>name</td><td>string</td><td>Name of an interface</td></tr>
<tr><td>kwargs</td><td>dict() string:int</td><td>Name and bitwidth of each port in an interface. Prefix (i_, o_, io_) should not be added.</td></tr>
</table>
</div>

<br>

<div class = "funcbox">
<h3>Interface.add_mode_port(name, *args)</h3>
<p>
This function defines a mode (directions) of port connecting to a block.
</p>
<table>
<tr><td>name</td><td>string</td><td>Name of a mode port</td></tr>
<tr><td>args</td><td>string</td><td>Name of ports and directions in a mode port. Prefix (i_, o_, io_) is needed to identify port direction.</td></tr>
</table>
</div>
